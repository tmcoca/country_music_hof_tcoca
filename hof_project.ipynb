```python
print(soup.prettify())
```

```python
members = soup.find_all('h3')
members = [str(x.text) for x in members]
members
```

```python
dates = soup.find_all(text = re.compile('Inducted'))
dates = [str(x.string) for x in dates]
dates = [re.search('\d{4}', x).group(0) for x in dates]
dates
```

```python
df = pd.DataFrame(dates, members)
df
```

```python
pages = []
URL = 'https://countrymusichalloffame.org/hall-of-fame/members/page/'
for number in range(1,17):
    new_URL = URL + str(number)
    pages.append(new_URL)
```

```python
total_members = []
total_dates = []
for URL in pages:
    response = requests.get(URL)
    soup = BS(response.text)
    members = soup.find_all('h3')
    members = [str(x.text) for x in members]
    for x in members:
        total_members.append(x)
    dates = soup.find_all(text = re.compile('Inducted'))
    dates = [str(x.string) for x in dates]
    dates = [re.search('\d{4}', x).group(0) for x in dates]
    for x in dates:
        total_dates.append(x)
```

```python
df = pd.DataFrame(total_dates, total_members)
```

```python
df.reset_index()
```

```python
df.reset_index().rename(columns = {'index':'Members', 0:'Year Inducted'})
```
